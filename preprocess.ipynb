```python
import pandas as pd
from pathlib import Path
cxr8_data_path = Path('D:\\') / 'data' / 'cxr8'
df = pd.read_csv(cxr8_data_path / 'Data_Entry_2017_v2020.csv')
df
```

```python
import shutil

for img_name in df[df['Finding Labels'].str.contains('No Finding')]['Image Index']:
    src = cxr8_data_path / 'no_finding' / img_name
    dst = cxr8_data_path / 'filtered' / img_name
    shutil.copyfile(src, dst)
```
